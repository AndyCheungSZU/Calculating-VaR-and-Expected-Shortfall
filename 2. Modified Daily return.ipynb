```python
import numpy as np
import pandas as pd
import datetime
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
Daily_return = pd.read_excel('/Users/andy/Desktop/PM/Daily_return.xlsx')
MSCI = pd.read_excel('/Users/andy/Downloads/MSCI World Price Index USD.xlsx')
```

```python
Daily_return.index = pd.to_datetime(Daily_return.index)
Daily_return = Daily_return.set_index('Date')
Daily_return.info()
```

```python
MSCI = MSCI.set_index('Date')
MSCI.index = pd.to_datetime(MSCI.index)
MSCI_5y = MSCI[(MSCI.index >= pd.to_datetime('20150101')) & (MSCI.index <= pd.to_datetime('20191231'))]
MSCI_5y.info()
```

```python
# MSCI_5y.to_csv('MSCI_5y.csv')
```

```python
df = pd.concat([Daily_return, MSCI_5y], axis=1)
df.head()
```

```python
df.drop(['Close'], axis=1, inplace=True)
df.rename(columns={'Return': 'MSCI'}, inplace=True)
df.fillna(0, inplace=True)
df.info()
```

```python
# df.to_csv('Matched Daily Return.csv')
```
